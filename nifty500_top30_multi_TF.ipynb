```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

# from google.colab import drive
# drive.mount('/content/drive')

# filepath = r"/content/drive/MyDrive/Quantzaar-tradingS, InvestmentS, Trade&investMyTheory/Backtesting/ind_nifty500list.csv"
# df=pd.read_csv(filepath)
# nifty500_symbols = df.Symbol + '.NS'

# Step 1: Get the NIFTY 500 stock symbols
nifty500_url = "https://en.wikipedia.org/wiki/NIFTY_500"
nifty500_symbols = pd.read_html(nifty500_url)[2][3][1:] + '.NS'

# Step 2: Download historical data for the NIFTY 500 stocks
df = yf.download(nifty500_symbols.to_list(), start='2022-01-01', end=datetime.now().strftime('%Y-%m-%d'))['Adj Close']

```

```python
# Step 3: Resample the data to monthly frequency and calculate the returns
df_monthly = df.resample('M').ffill()  # Resample to monthly frequency
df_12mo_ret = df_monthly.pct_change(12)  # 12-month returns (1-year return)
#it will calculate feb_start/jan_end to jan end for date 31jan
df_6mo_ret = df_monthly.pct_change(6)    # 6-month returns
df_3mo_ret = df_monthly.pct_change(3)    # 3-month returns

# Step 4: Get the latest available dates (last two months)
#in between months we will consider last months' data.
latest_dates = df_12mo_ret.index[-2:]  # Last two months
print(f"\nLatest Dates: {latest_dates}")

# Step 5: Get top 30 stocks based on the latest 1-year, 6-month, and 3-month returns for the last two months
results = {}
for date in latest_dates:
    top30_1year = df_12mo_ret.loc[date].nlargest(30).index
    top30_6month = df_6mo_ret.loc[date].nlargest(30).index
    top30_3month = df_3mo_ret.loc[date].nlargest(30).index

    results[date] = {
        '1-Year Top 30 Stocks': top30_1year,
        '6-Month Top 30 Stocks': top30_6month,
        '3-Month Top 30 Stocks': top30_3month
    }

    # Display the top 30 lists
    print(f"\nTop 30 Stocks for Date {date} Based on 1-Year Return:")
    print(top30_1year)

    print(f"\nTop 30 Stocks for Date {date} Based on 6-Month Return:")
    print(top30_6month)

    print(f"\nTop 30 Stocks for Date {date} Based on 3-Month Return:")
    print(top30_3month)

# Step 6: Save the top 30 stocks for each period into an Excel file
# Specify the desired path to save the file
file_path = r"/content/drive/MyDrive/Quantzaar-tradingS, InvestmentS, Trade&investMyTheory/Backtesting/top_30_stocks.xlsx"

with pd.ExcelWriter(file_path) as writer:
    for date, data in results.items():
        pd.DataFrame(np.array(data['1-Year Top 30 Stocks']), columns=['1-Year Top 30 Stocks']).to_excel(writer, sheet_name=f"Top30_1Year_{date.strftime('%Y-%m')}", index=False)
        pd.DataFrame(np.array(data['6-Month Top 30 Stocks']), columns=['6-Month Top 30 Stocks']).to_excel(writer, sheet_name=f"Top30_6Month_{date.strftime('%Y-%m')}", index=False)
        pd.DataFrame(np.array(data['3-Month Top 30 Stocks']), columns=['3-Month Top 30 Stocks']).to_excel(writer, sheet_name=f"Top30_3Month_{date.strftime('%Y-%m')}", index=False)

print(f"The top 30 stocks for the last two months have been saved to '{file_path}'.")
```
